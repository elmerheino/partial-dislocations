```python
import matplotlib.pyplot as plt
import glob
import pandas as pd
import numpy as np
import pickle
from pathlib import Path

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
p = Path("../results/11-10-pieni-c_gamma/partial/l-128-d-8/shape_data/extra_info_dump-l-128-s-2-d0-8.pickle")
with open(p, "rb") as fp:
    data = pickle.load(fp)

df_noise = pd.read_csv("../results/11-10-pieni-c_gamma/partial/l-128-d-8/force_data/noise_tauc_data_l-128-s-2-d0-8.csv")

index = 13

deltaR = df_noise['deltaR'][index]

fig, axes = plt.subplots(figsize=(12, 4))

# Plot convergence status
convergence = data[index]['converged']
tau_ext_values = data[index]['tau_ext']  # Assuming all data points have the same tau_ext
shape = data[index]['shapes']
y1, y2 = shape[0]

print(np.mean((y1-y2)))

axes.plot(tau_ext_values, convergence, marker='o')
axes.set_title('Convergence Status')
axes.set_xlabel('$\\tau_{ext}$')
axes.set_ylabel('Converged')
axes.set_yticks([0, 1])
axes.set_yticklabels(['False', 'True'])
axes.set_title(f'L = 128, $\Delta R = 10^{{{np.log10(deltaR)}}}$')
axes.grid(True)

plt.tight_layout()
plt.show()
```

```python
def make_d0_plot(ax, index):
    p = Path("../results/11-10-pieni-c_gamma/partial/l-128-d-64/shape_data/extra_info_dump-l-128-s-2-d0-64.pickle")

    with open(p, "rb") as fp:
        data = pickle.load(fp)

    df_noise = pd.read_csv("../results/11-10-pieni-c_gamma/partial/l-128-d-64/force_data/noise_tauc_data_l-128-s-2-d0-64.csv")

    deltaR = df_noise['deltaR'][index]
    d = data[index]

    converged_shapes = [i for i,j in zip(d['shapes'], d['converged']) if j==True]
    converged_taus = [i for i,j in zip(d['tau_ext'], d['converged']) if j==True]

    d_rel = [np.mean(y[0] - y[1]) for y in converged_shapes]

    plt.scatter(converged_taus, d_rel, marker='.')

    ax.set_title(f'$\Delta R = 10^{{{ np.log10(deltaR) :.3f}}}$, L=128')
    ax.grid(True)
    ax.set_yscale('log')

    ax.set_xlabel('$\\tau_{ext}$')
    ax.set_ylabel('$d_{rel}$')

    ax.axhline(y=64, color='r', linestyle='--', label="$d_0$ = 64")

    ax.legend()

fig, ax = plt.subplots(figsize=(4,4))
make_d0_plot(ax, 7)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
make_d0_plot(ax, 0)
```
